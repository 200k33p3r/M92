```python
dp = pd.read_csv('C:/Users/marti/Desktop/school work/Dartmouth/M92 project/M92_fits_with_err_with_loc')
df = dp.copy()
df['vv'] = df['vv'] - 17.925 + (-13.77)
df['vi'] = df['vi'] - 0.432 +(-0.375)
df['ii'] = df['vv'] - df['vi']
```

```python
i_bin = []
for i in range(len(df)):
    if df['r'].values[i] < Rad_range[0] and df['ii'].values[i] < Mag_range[0]:
        i_bin.append(0)
    elif df['r'].values[i] < Rad_range[0]:
        m_idx = 0
        found=False
        while m_idx < 7 and found==False:
            if df['ii'].values[i] > Mag_range[m_idx+1] or df['ii'].values[i] < Mag_range[m_idx]:
                m_idx += 1
            else:
                found=True
                i_bin.append((m_idx+1)*10)
        if found==False:
            i_bin.append(70)
    elif df['ii'].values[i] < Mag_range[0]:
        r_idx = 0
        found=False
        while r_idx < 9 and found==False:
            if df['r'].values[i] > Rad_range[r_idx+1] or df['r'].values[i] < Rad_range[r_idx]:
                r_idx += 1
            else:
                found=True
                i_bin.append((r_idx+1))
        if found==False:
            i_bin.append(9)
    else:
        m_idx = 0
        r_idx = 0
        found_m=False
        found_r=False
        while m_idx < 7 and found_m==False:
            if df['ii'].values[i] > Mag_range[m_idx+1] or df['ii'].values[i] < Mag_range[m_idx]:
                m_idx += 1
            else:
                found_m=True
        while r_idx < 9 and found_r==False:
            if df['r'].values[i] > Rad_range[r_idx+1] or df['r'].values[i] < Rad_range[r_idx]:
                r_idx += 1
            else:
                found_r=True
        if found_m==True and found_r==True:
            i_bin.append((r_idx+1+(m_idx+1)*10))
        elif found_m==True:
            i_bin.append((8+1+(m_idx+1)*10))
        elif found_r==True:
            i_bin.append((r_idx+1+(6+1)*10))
        else:
            i_bin.append(79)
```
